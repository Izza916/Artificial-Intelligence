Task#01:

 Write Python Program for each task, create sample space and calculate the
probability and show the output.
A manufacturing firm employs three analytical plans for the design and development of
a particular product. For cost reasons, all three are used at varying times. In fact, plans
1, 2, and 3 are used for 30%, 20%, and 50% of the products, respectively. The defect rate is different for the three procedures as follows: P(D|P1)=0.01, P(D|P2)=0.03,
P(D|P3)=0.02, where P(D|Pj ) is the probability of a defective product, given plan j. If a random product was observed and found to be defective, which plan was most likely used and thus responsible?

```python
#p(d|p)= p(A) * P()
def posterior_probability(defect_probabilities, plan_probabilities):
    posterior_numerators = [defect_probabilities[i] * plan_probabilities[i] for i in range(len(defect_probabilities))]
    total_defect_probability = sum(posterior_numerators)
    posterior_probabilities = [numerator / total_defect_probability for numerator in posterior_numerators]
    return posterior_probabilities

def main():
    defect_probabilities = [0.01, 0.03, 0.02]
    plan_probabilities = [0.30, 0.20, 0.50]
    posterior_probabilities = posterior_probability(defect_probabilities, plan_probabilities)
    most_likely_plan_index = posterior_probabilities.index(max(posterior_probabilities))
    most_likely_plan = most_likely_plan_index + 1
    print("Posterior probabilities for each plan:", posterior_probabilities)
    print("Most likely plan responsible for the defective product:", most_likely_plan)

if __name__ == "__main__":
    main()
```

Task#02:

Sarah is a plant manager at a manufacturing facility. She&#39;s concerned about two
potential issues that could affect production: a malfunction in one of the machines or a
power outage in the area. Sarah relies on two employees, Alex and Emily, to notify her
if either event occurs. Alex is diligent and always calls Sarah when he notices a
machine malfunction, but occasionally he confuses the sound of the machine with other
noises in the factory and calls mistakenly. Emily, on the other hand, tends to get
engrossed in her work and sometimes misses the sound of the machine
malfunctioning. Given this scenario, we want to calculate the probability of a Machine
Malfunction Alarm.

```python
# pomogranate method #01
# variables nodes
# bayesiun network
# nodes
# create and add in bayesiun network
# add states
# add edges
#------- machine--------
#     /         \
#    /            \
#   emily        alex
#       \         /
#        \       /
#         sarah
from pomegranate import *
alex = DiscreteDistribution({'call': 0.9, 'no_call': 0.1})
emily = DiscreteDistribution({'call': 0.8, 'no_call': 0.2})
machine_alarm = ConditionalProbabilityTable(
    [['call', 'call', 'call', 0.95],
     ['call', 'call', 'no_call', 0.05],
     ['call', 'no_call', 'call', 0.7],
     ['call', 'no_call', 'no_call', 0.3],
     ['no_call', 'call', 'call', 0.05],
     ['no_call', 'call', 'no_call', 0.95],
     ['no_call', 'no_call', 'call', 0.3],
     ['no_call', 'no_call', 'no_call', 0.9]],
    [alex, emily]
)
s1 = State(alex, name="Alex")
s2 = State(emily, name="Emily")
s3 = State(machine_alarm, name="MachineAlarm")
network = BayesianNetwork("Machine Malfunction Alarm")
network.add_states(s1, s2, s3)
network.add_edge(s1, s3)
network.add_edge(s2, s3)
network.bake()
probability = network.predict_proba({})
for state, prob in zip(network.states, probability):
    if state.name == 'MachineAlarm':
        print("Probability of Machine Malfunction Alarm:", prob.parameters[0]['call'])
```

Task#03:

Suppose there are three boxes, each containing two coins. Box 1 has two
gold coins, Box 2 has two silver coins, and Box 3 has one gold coin and one silver coin.
You randomly pick a box and take out one coin. It turns out to be a gold coin. What is the
probability that the other coin in the box is also gold if:

(a) You picked Box 1?
(b) You picked Box 2?
(c) You picked Box 3?

Use ‘pomegranate’ and create conditional probability table for each box.

```python
from pomegranate import *

box1 = DiscreteDistribution({'gold': 1, 'silver': 0})
box2 = DiscreteDistribution({'gold': 0, 'silver': 1})
box3 = DiscreteDistribution({'gold': 0.5, 'silver': 0.5})

table_box1 = ConditionalProbabilityTable(
    [['gold', 'gold', 1.0],
     ['gold', 'silver', 0.0],
     ['silver', 'gold', 0.0],
     ['silver', 'silver', 1.0]],
    [box1]
)

table_box2 = ConditionalProbabilityTable(
    [['gold', 'gold', 0.0],
     ['gold', 'silver', 1.0],
     ['silver', 'gold', 1.0],
     ['silver', 'silver', 0.0]],
    [box2]
)

table_box3 = ConditionalProbabilityTable(
    [['gold', 'gold', 0.5],
     ['gold', 'silver', 0.5],
     ['silver', 'gold', 0.5],
     ['silver', 'silver', 0.5]],
    [box3]
)

state_box1 = State(box1, name='box1')
state_box2 = State(box2, name='box2')
state_box3 = State(box3, name='box3')

state_table_box1 = State(table_box1, name='table_box1')
state_table_box2 = State(table_box2, name='table_box2')
state_table_box3 = State(table_box3, name='table_box3')

network = BayesianNetwork("Coin Selection")

network.add_states(state_box1, state_box2, state_box3,
                    state_table_box1, state_table_box2, state_table_box3)

network.add_edge(state_box1, state_table_box1)
network.add_edge(state_box2, state_table_box2)
network.add_edge(state_box3, state_table_box3)

network.bake()

observations = {'table_box1': 'gold'}
prob_box1_other_gold = network.predict_proba(observations)[0].parameters[0]['gold']

observations = {'table_box2': 'gold'}
prob_box2_other_gold = network.predict_proba(observations)[1].parameters[0]['gold']

observations = {'table_box3': 'gold'}
prob_box3_other_gold = network.predict_proba(observations)[2].parameters[0]['gold']

print("Probability of the other coin being gold if Box 1 is picked:", prob_box1_other_gold)
print("Probability of the other coin being gold if Box 2 is picked:", prob_box2_other_gold)
print("Probability of the other coin being gold if Box 3 is picked:", prob_box3_other_gold)
```

Task#04:

Scenario:
A company, &quot;Smart Energy Solutions,&quot; specializes in providing renewable energy
solutions to households. They offer two types of energy products: solar panels and
wind turbines. The demand for these products can fluctuate based on various factors
such as weather conditions, government incentives, and consumer preferences. The
company wants to use a Hidden Markov Model (HMM) to predict the demand state of
each energy product based on historical sales data.
Task:
The company has collected sales data for a week for both types of energy products:
solar panels and wind turbines. Based on this data, they aim to train an HMM to predict
the demand state of each product (high-demand or low-demand) and determine the
transition and emission probabilities.

```python
!pip install hmmlearn
```

```python
import numpy as np
from hmmlearn import hmm

solar_panel_sales = [20, 25, 30, 15, 10, 35, 40]
wind_turbine_sales = [10, 15, 20, 25, 30, 20, 15]
sales_data = np.column_stack([solar_panel_sales, wind_turbine_sales])
model = hmm.MultinomialHMM(n_components=2, n_iter=100)
model.transmat_ = np.array([[0.5, 0.5], [0.5, 0.5]])

model.fit(sales_data)
print("Transition matrix:")
print(model.transmat_)
print("\nEmission probabilities (solar panels):")
print(model.emissionprob_[0])
print("\nEmission probabilities (wind turbines):")
print(model.emissionprob_[1])
new_sales_data = np.array([[30, 20]])

predicted_states = model.predict(new_sales_data)

demand_states = ["low-demand", "high-demand"]
predicted_demand_state = demand_states[predicted_states[0]]

print("Predicted demand state:", predicted_demand_state)
```

Task#05:

Scenario:
In a certain town, the weather can either be sunny or rainy. People in this town dress
differently based on the weather. On sunny days, they typically wear t-shirts, while on
rainy days, they wear jackets to stay dry.
Task:
We have a record of what people wear for a week but don&#39;t have direct observations of
the weather. Based on the clothing worn each day, use a Hidden Markov Model to infer
the most likely sequence of weather states (sunny or rainy) for the week.

```python
import numpy as np
from hmmlearn import hmm

transition_matrix = np.array([[0.7, 0.3],
                              [0.4, 0.6]])

emission_matrix = np.array([[0.9, 0.1],
                             [0.2, 0.8]])

start_probabilities = np.array([0.5, 0.5])

model = hmm.MultinomialHMM(n_components=2)
model.startprob_ = start_probabilities
model.transmat_ = transition_matrix
model.emissionprob_ = emission_matrix

observations = np.array([[0], [1], [1], [0], [0], [1], [0]])
observations = observations.reshape(-1, 1)
model.fit(observations)
most_likely_states = model.predict(observations)

weather_map = {0: 'Sunny', 1: 'Rainy'}
predicted_weather = [weather_map[state] for state in most_likely_states]
print("Most likely sequence of weather states for the week:")
print(predicted_weather)
```
