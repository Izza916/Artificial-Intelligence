Task#01: Write Python Program for each task, create sample space and calculate the probability and show the output.A manufacturing firm employs three analytical plans for the design and development of a particular product. For cost reasons, all three are used at varying times. In fact, plans 1, 2, and 3 are used for 30%, 20%, and 50% of the products, respectively.

The defect Crate is different for the three procedures as follows:

P(D|P1)=0.01,

P(D|P2)=0.03

P(D|P3)=0.02

where P(D|Pj ) is the probability of a defective product, given plan j. If a
random product was observed and found to be defective, which plan was most likely used and thus responsible?

```python
#p(d|p)= p(A) * P()
def posterior_probability(defect_probabilities, plan_probabilities):
    posterior_numerators = [defect_probabilities[i] * plan_probabilities[i] for i in range(len(defect_probabilities))]
    total_defect_probability = sum(posterior_numerators)
    posterior_probabilities = [numerator / total_defect_probability for numerator in posterior_numerators]
    return posterior_probabilities

def main():
    defect_probabilities = [0.01, 0.03, 0.02]
    plan_probabilities = [0.30, 0.20, 0.50]
    posterior_probabilities = posterior_probability(defect_probabilities, plan_probabilities)
    most_likely_plan_index = posterior_probabilities.index(max(posterior_probabilities))
    most_likely_plan = most_likely_plan_index + 1
    print("Posterior probabilities for each plan:", posterior_probabilities)
    print("Most likely plan responsible for the defective product:", most_likely_plan)

if __name__ == "__main__":
    main()
```

Task#02:
Sarah is a plant manager at a manufacturing facility. She&#39;s concerned about two
potential issues that could affect production: a malfunction in one of the machines or a
power outage in the area. Sarah relies on two employees, Alex and Emily, to notify her
if either event occurs. Alex is diligent and always calls Sarah when he notices a
machine malfunction, but occasionally he confuses the sound of the machine with other
noises in the factory and calls mistakenly. Emily, on the other hand, tends to get
engrossed in her work and sometimes misses the sound of the machine
malfunctioning. Given this scenario, we want to calculate the probability of a Machine
Malfunction Alarm.

```python
# pomogranate method #01
# variables nodes
# bayesiun network
# nodes
# create and add in bayesiun network
# add states
# add edges
```

```python
#------- machine--------
#     /         \
#    /            \
#   emily        alex
#       \         /
#        \       /
#         sarah
from pomegranate import *
alex = DiscreteDistribution({'call': 0.9, 'no_call': 0.1})
emily = DiscreteDistribution({'call': 0.8, 'no_call': 0.2})
machine_alarm = ConditionalProbabilityTable(
    [['call', 'call', 'call', 0.95],
     ['call', 'call', 'no_call', 0.05],
     ['call', 'no_call', 'call', 0.7],
     ['call', 'no_call', 'no_call', 0.3],
     ['no_call', 'call', 'call', 0.05],
     ['no_call', 'call', 'no_call', 0.95],
     ['no_call', 'no_call', 'call', 0.3],
     ['no_call', 'no_call', 'no_call', 0.9]],
    [alex, emily]
)
s1 = State(alex, name="Alex")
s2 = State(emily, name="Emily")
s3 = State(machine_alarm, name="MachineAlarm")
network = BayesianNetwork("Machine Malfunction Alarm")
network.add_states(s1, s2, s3)
network.add_edge(s1, s3)
network.add_edge(s2, s3)
network.bake()
probability = network.predict_proba({})
for state, prob in zip(network.states, probability):
    if state.name == 'MachineAlarm':
        print("Probability of Machine Malfunction Alarm:", prob.parameters[0]['call'])
```
