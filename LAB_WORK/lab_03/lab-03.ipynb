# Task#01: Simple Reflex Agents
Consider an interactive and cognitive environment (ICE) in which a smart camera is
monitoring robot movement from one location to another. Let a robot be at location A for
some time instant and then moves to point B and eventually reaches at point C and so on
and so forth shown in the Fig. Develop a Python code to calculate a distance between
reference point R (4, 0) of a camera and A, B, and C and N number of locations.

```python
import math

class RobotEnvironment:
    def __init__(self, rows, columns):
        self.rows = rows
        self.columns = columns
        self.camera_location = (4, 0)

class SimpleReflexAgent:
    def __init__(self, environment):
        self.location = (0, 0)
        self.environment = environment

    def move_to(self, new_location):
        self.location = new_location

    def calculate_distance(self, location):
        return math.sqrt((self.environment.camera_location[0] - location[0])**2 +
                         (self.environment.camera_location[1] - location[1])**2)

    def next_action(self, location_label):
        distance_to_location = self.calculate_distance(self.location)
        print(f"Distance to {location_label}: {distance_to_location}")

def main():
    environment = RobotEnvironment(11, 19)
    agent = SimpleReflexAgent(environment)

    locations = {
        "A": (0, 0),
        "B": (0, 9),
        "C": (0, 18)
    }

    for location_label, location_coords in locations.items():
        agent.move_to(location_coords)
        print(f"\nRobot at location {location_label}:")
        agent.next_action(location_label)

if __name__ == "__main__":
    main()
```

# Task#02 Simple Reflex Agents

Consider a scenario, cameras placed on every side of the car — front, rear, left and right — to stitch together a 360-degree view of the environment. For a three-lane road a car is moving on a middle lane, consider the below scenario

 If the front camera detects the object within range of 8 meters breaks are applied automatically.

 If the left camera detects the object within range of 2 meters car moves to the right lane.

 If the right camera detects the object within range of 2 meters car moves to the left lane.

 For parking the car if the rear camera detects the object within 5 cm breaks are applied.

```python
class CarEnvironment:
    def __init__(self):
        self.front_camera_range = 8
        self.side_camera_range = 2
        self.rear_camera_range = 0.05

        self.car_position = "middle"

    def percept(self):
        return {
            "front": self.front_camera_range,
            "left": self.side_camera_range,
            "right": self.side_camera_range,
            "rear": self.rear_camera_range
        }

    def act(self, action):
        if action == "move_left":
            self.car_position = "left"
        elif action == "move_right":
            self.car_position = "right"
        elif action == "apply_brakes":
            print("Brakes applied!")

class SimpleReflexAgent:
    def program(self, percept):
        if percept["rear"] <= 0.05:
            return "apply_brakes"
        elif percept["front"] <= 8:
            return "apply_brakes"
        elif percept["left"] <= 2 and percept["front"] > 8:
            return "move_right"
        elif percept["right"] <= 2 and percept["front"] > 8:
            return "move_left"
        else:
            return "do_nothing"

env = CarEnvironment()
agent = SimpleReflexAgent()
for _ in range(10):
    percept = env.percept()
    action = agent.program(percept)
    env.act(action)
```
