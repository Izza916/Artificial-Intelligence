Q5) Perform the following tasks with NLTK.
1. Write a Python NLTK program to split the text sentence/paragraph into a listof words.
Sample Below.

```python
import nltk
nltk.download('punkt')
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

text = "Joe waited for the train. The train was late. Mary and Samantha took the bus. I looked for Mary and Samantha at the bus station."
sentences = sent_tokenize(text)
words_list = [word_tokenize(sentence) for sentence in sentences]
print("Original string:")
print(text)
print("Sentence-tokenized copy in a list:")
print(sentences)
print("Words list:")
for words in words_list:
    print(words)
```

3. Write a Python NLTK program to tokenize words, sentence wise.

```python
import nltk

def tokenize_sentence_wordwise(text):
    sentences = nltk.sent_tokenize(text)
    tokenized_sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
    return tokenized_sentences

original_string = "Joe waited for the train. The train was late. Mary and Samantha took the bus. I looked for Mary and Samantha at the bus station."
tokenized_words_sentencewise = tokenize_sentence_wordwise(original_string)
print("Tokenize words sentence wise:")
for sentence in tokenized_words_sentencewise:
    print("Read the list:")
    print(sentence)
```

```python
import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm")
def visualize_dependency(sentence):
    doc = nlp(sentence)
    displacy.render(doc, style="dep", options={'distance': 100})

sentence = "The quick brown fox jumps over the lazy dog."
visualize_dependency(sentence)
```

2. Break a given sentence into tokens each representing a single word.

```python
import spacy

nlp = spacy.load("en_core_web_sm")
def tokenize_sentence(sentence):
    doc = nlp(sentence)
    tokens = [token.text for token in doc]
    return tokens

sentence = "The quick brown fox jumps over the lazy dog."
print("Tokenized sentence:", tokenize_sentence(sentence))
```
